```python
annotation_count_df = pd.DataFrame(columns=[0, 1, 2])
for file_name, temp_data in data.groupby(["File"]):
    temp_object = pd.DataFrame(temp_data['Annotation'].value_counts()).T
    temp_object.index = [file_name]
    annotation_count_df = pd.concat([annotation_count_df, temp_object], axis=0)

annotation_count_df.fillna(0, inplace=True)
annotation_count_df
```
